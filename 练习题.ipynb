```python
# 斐波那契数列。程序分析：斐波那契数列（Fibonacci sequence），又称黄金分割数列，指的是这样一个数列：0、1、1、2、3
# 在数学上，费波那契数列是以递归的方法来定义：

# 斐波拉契递归
def fib(n):
    if n == 1:
        return 0
    if n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
fib(5)
    
```
